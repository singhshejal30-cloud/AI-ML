```python
import pandas as pd
```

```python
df=pd.read_csv("customer_segmentation_realistic.csv")
df
```

```python
from sklearn.cluster import KMeans
```

```python
KM=KMeans(n_clusters=4, random_state=42) #n_cluster for in chart colors
df["Cluster"]=KM.fit_predict(df)
```

```python
print(df.head())
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(df['AnnualIncome'],df['SpendingScore'],c=df['Cluster'])
plt.xlabel("Annual Income")  
plt.ylabel("Spending")
plt.title("customer_segmentation")
plt.show()
```

```python
df=pd.read_csv("student_marks_clustering.csv")
df
```

```python
X=df[["Total"]]
```

```python
KMs=KMeans(n_clusters=3 , random_state=42)
df['Cluster']=KMs.fit_predict(X)
```

```python
plt.scatter(df['StudentID'],df['Total'],c=df['Cluster'])
plt.xlabel("Std ID")  
plt.ylabel("Total Marks")
plt.title("Students Performance Cluster")
plt.show()
```
