```python
import numpy as np
import pandas as pd
from scipy import stats
```

```python
#series
data = [70,90,65,80,85]
print("mean",np.mean(data))
```

```python
print("median",np.median(data))
```

```python
print("Most frequent value",stats.mode(data))
```

```python
print(max(data)-min(data))
```

```python
print("Variance",np.var(data))
```

```python
print("SD",np.std(data))
```
