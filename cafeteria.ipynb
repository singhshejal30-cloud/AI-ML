```python
import pandas as pd
import numpy as np
from scipy import stats
```

```python
Spending =[45, 60, 40, 55, 70, 65, 50, 45, 90, 85, 60, 55, 50, 45, 70, 75, 65, 50, 55, 60, 80, 85, 75, 60, 55, 45, 70, 65, 60, 55,115]
```

```python
df=pd.DataFrame(Spending, columns = ["Spending"])
```

```python
print("Mean",np.mean(Spending))
```

```python
print("Median",np.median(Spending))
```

```python
print("Most frequent value",stats.mode(Spending))
```

```python
print("Variance",np.var(Spending))
```

```python
print("SD",np.std(Spending))
```

```python
xyz = range(40, 110, 10)

frequency_table = pd.cut(df["Spending"], bins=xyz).value_counts().sort_index()
frequency_table
```

```python
import matplotlib.pyplot as plt
```

```python
plt.hist(Spending, bins=xyz, edgecolor ="black")
plt.title("Histogram of Cafeteria Spending")
plt.xlabel("Spending")
plt.ylabel("Frequency")
plt.show()
```

```python
Q1 = df["Spending"].quantile(0.25)
Q3 = df["Spending"].quantile(0.75)
IQR = Q3 - Q1

```

```python
IQR #how much middle data spread with mean value
```

```python
#Range -if any value cross it that an outliers
x = Q1 - 1.5 * IQR
y = Q3 + 1.5 * IQR
```

```python
outliers = df[(df["Spending"]<x) | (df["Spending"]>y)]
```

```python
heights = [160, 165, 170, 168, 172, 169, 171, 167, 166, 173,
           175, 169, 168, 170, 172, 174, 165, 167, 168, 171]

height = pd.DataFrame(heights)
```

```python
mean_height = height.mean()
SD = height.std()

cv = (SD / mean_height) * 100


print("Mean Height:", round(mean_height,2))
print("Standard Deviation:", round(SD,2))
print("Coefficient of Variation (CV):", round(cv,2), "%")
```
