```python
mean_marks=df[["Math","Science","English"]].mean()
```

```python
print(mean_marks)
```

```python
median_marks=df[["Math","Science","English"]].median()
```

```python
print(median_marks)
```

```python
mode_marks=df[["Math","Science","English"]].mode()
print(mode_marks)
```

```python
max_marks=df[["Math","Science","English"]].max()
print(max_marks)
```

```python
min_marks=df[["Math","Science","English"]].min()
print(min_marks)
```

```python
df["Total"] = df[["Math","Science","English"]].sum(axis=1)
```

```python
print(df)
```

```python
highest = df.loc[df["Total"].idxmax()]
print("Highest Scorer:")
print(highest)
```

```python
lowest = df.loc[df["Total"].idxmin()]
print("lowest Scorer:")
print(lowest)
```
